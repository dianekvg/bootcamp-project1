```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
file_name1 = os.path.join("winter.csv")
file_name2 = os.path.join("olympics_countries.csv")
file_name3 = os.path.join("dictionary.csv")
file_name4 = os.path.join("Total_Population.csv")
file_name5 = os.path.join("Regional_Population.csv")

winter_df = pd.read_csv(file_name1)
olympic_countries_df = pd.read_csv(file_name2)
dictionary_df = pd.read_csv(file_name3)
population_df = pd.read_csv(file_name4, skiprows=4)
regions_df = pd.read_csv(file_name5)
```

```python
winter_df.head()
```

```python
new_winter_df = winter_df.rename(index=str, columns={"Country": "Code"})
```

```python
new_winter_df.head()
```

```python
winter_olympic_df = new_winter_df.merge(dictionary_df, how="outer", on="Code")
```

```python
winter_olympic_df.head(
)
```

```python
olympic_countries_df.head()
```

```python
winter_olympics_df = winter_olympic_df.merge(olympic_countries_df, how="outer", on="Country")
```

```python
winter_olympics_df.head()
```

```python
winter_olympics = winter_olympics_df.drop(["Rk"], axis=1)
winter_olympics.head()
```

```python
wpop= winter_olympics.groupby("Country").Population.mean()
wpop.head()
```

```python
wt= winter_olympics.groupby("Country").Winter_Total.mean()
wt.head()
```

```python
wp = winter_olympics.groupby("Country").Winter_Participants.mean()
wp.head()
```

```python
wg = winter_olympics.groupby("Country").Winter_Gold.mean()
ws = winter_olympics.groupby("Country").Winter_Silver.mean()
wb = winter_olympics.groupby("Country").Winter_Bronze.mean()
```

```python
Scatter_Table = pd.DataFrame({"Country Population": wpop,
                              "Total Medals Won": wt,
                              "Total Athlete Participants": wp,
                              "Gold Medals": wg,
                              "Silver Medals": ws,
                              "Bronze Medals": wb    
                             })
Scatter_Table.head()
Scatter_Table.fillna(0, inplace=True)
Scatter_Table.head()
```

```python
Country = Scatter_Table[Scatter_Table["Total Athlete Participants"] != 0]
```

```python
Country.to_csv("Countries_with_athletes.csv")
```

```python
gold_medals = winter_olympic_df.loc[winter_olympic_df["Medal"] == "Gold"]
silver_medals = winter_olympic_df.loc[winter_olympic_df["Medal"] == "Silver"]
bronze_medals = winter_olympic_df.loc[winter_olympic_df["Medal"] == "Bronze"]
```

```python
USAgold_medals = gold_medals.loc[gold_medals["Code"] == "USA"]
USAsilver_medals = silver_medals.loc[silver_medals["Code"] == "USA"]
USAbronze_medals = bronze_medals.loc[bronze_medals["Code"] == "USA"]
```

```python
gold_bar = USAgold_medals.groupby("Year").Medal.count()
silver_bar = USAsilver_medals.groupby("Year").Medal.count()
bronze_bar = USAbronze_medals.groupby("Year").Medal.count()
```

```python
Bar_Table = pd.DataFrame({"Gold_Medals": gold_bar,
                              "Silver_Medals": silver_bar,
                              "Bronze_Medals": bronze_bar,

                        })

Bar_Table.head()
Bar_Table.fillna(0, inplace=True)
Bar_Table.head()
```
